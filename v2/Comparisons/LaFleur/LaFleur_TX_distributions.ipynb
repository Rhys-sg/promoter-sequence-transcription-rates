```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../Data/Combined/LaFleur_supp.csv')
df.head()
```

```python
# Graph the distribution of 'Observed log(TX/Txref)' using a kdensity plot

sns.kdeplot(data=df, x='Observed log(TX/Txref)', fill=True, common_norm=False, alpha=0.5)
plt.title('Distribution of Observed log(TX/Txref) by File Name')
plt.xlabel('Observed log(TX/Txref)')
plt.ylabel('Density')
plt.legend(title='File Name')
plt.show()
```

```python
# Graph the distribution of 'Observed log(TX/Txref)' using a kdensity plot, seperated by 'File Name' (color)

sns.kdeplot(data=df, x='Observed log(TX/Txref)', hue='File Name', fill=True, common_norm=False, alpha=0.5)
plt.title('Distribution of Observed log(TX/Txref) by File Name')
plt.xlabel('Observed log(TX/Txref)')
plt.ylabel('Density')
plt.legend(title='File Name')
plt.show()
```
