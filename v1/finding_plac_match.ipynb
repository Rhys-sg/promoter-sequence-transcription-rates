```python
def format(str):
    return str.replace(" ", "").upper()
```

```python
def differences(str1, str2):
    return [str1[i] == str2[i] for i in range(min(len(str1), len(str2)))].count(False) + abs(len(str1) - len(str2))
```

## Finding matches for pUC19.pdf

```python
# in order of priority (real order = h35, up, spacer, h10)

h10 = 'TATGTT'
h35 = 'TTGACA'
spacs = 'CTTTATGCTTCCGGCTCG'
up = 'ACTCATTAGGCACCCCAGGC' 
```

```python
# sequences with h10 match

pUC19_h10_df = df[(df['h10'] == h10)]
pUC19_h10_df.head()
```

```python
# sequences with h35 match

pUC19_h35_df = df[(df['h35'] == h35)]
pUC19_h35_df.head()
```

```python
# sequences with h10 and h35 match

pUC19_h10_h35_df = df[(df['h10'] == h10) & (df['h35'] == h35)]
pUC19_h10_h35_df.head()
```

## Finding matches for p5.pdf

```python
# This is in order of match priority and does not inlcude: -10 extended (not in LaFleur et al.), 
# discriminator (disc), or initial transcribed region (ITR).

unedited_h10 = 'T A A T T T'
unedited_h35 = 'T T G A C A'
unedited_spacs = 'C G A G C C G G A T G A T T A A T'
unedited_up = ''

h10 = format(unedited_h10)
h35 = format(unedited_h35)
spacs = format(unedited_spacs)
up = format(unedited_up)

print(f'Formatted h10: {h10}')
print(f'Formatted h35: {h35}')
print(f'Formatted spacs: {spacs}')
print(f'Formatted up: {up}')

print()

print(f'Length of h10: {len(h10)}')
print(f'Length of h35: {len(h35)}')
print(f'Length of spacs: {len(spacs)}')
print(f'Length of up: {len(up)}')
```
