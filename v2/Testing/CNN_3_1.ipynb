```python
import CNN_3_1 as parent
```

```python
name = 'CNN_3_1'
```

```python
# File path
filepath = '../Data/LaFleur_stacking_data.csv'

# Load and preprocess data
X, y, df, Prod_pred = parent.load_and_preprocess_data(filepath)
```

```python
# Encode sequences
X_encoded, max_length = parent.encode_sequences(X, Prod_pred)
```

```python
X_encoded[0]
```

```python
# Train-test split
import numpy as np

X_train, X_test, y_train, y_test = parent.train_test_split(X_encoded, y, test_size=0.2, random_state=42)

X_train = np.array([x[:-1] for x in X_train])
X_test = np.array([x[:-1] for x in X_test])
ProD_train = np.array([x[-1][0] for x in X_train])
ProD_test = np.array([x[-1][0] for x in X_test])

print(X_train.shape)
print(X_test.shape)
print(ProD_train.shape)
print(ProD_test.shape)
```

```python
# Only run if the model has not been trained before

# Build and train model
model = parent.build_model(input_shape=X_encoded.shape[1:])
history = parent.train_model(model, X_train, y_train, X_test, y_test)

# Save model
parent.save_model(model, name)
```

```python
# Load model if needed, if not, this will load the just-saved model

model = parent.load_saved_model(name)
```

```python
# Evaluate model
loss = parent.evaluate_model(model, X_test, y_test)
print(f'Loss: {loss}')
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

predictions = model.predict(X_train)

print("Predictions:", predictions.shape)
print("Prod_pred:", ProD_train.shape)

lin_regress_df = pd.DataFrame({
    'CNN': predictions.ravel(),
    'Prod_pred': ProD_train
})
y_train_df = pd.DataFrame({'y_train': y_train})

# Generate polynomial and interaction features
poly = PolynomialFeatures(degree=2)  # You can change the degree to the desired polynomial degree
lin_regress_poly = poly.fit_transform(lin_regress_df)

lin_reg = LinearRegression()

# Fit the model using polynomial features
lin_reg.fit(lin_regress_poly, y_train_df)

# Predict using the polynomial model
y_pred = lin_reg.predict(lin_regress_poly)

print("Predicted values:", y_pred)
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.scatter(lin_regress_df['CNN'], y_train, color='blue', label='y_train')
plt.scatter(lin_regress_df['CNN'], y_pred, color='red', label='y_pred', alpha=0.5)
plt.plot(lin_regress_df['CNN'], y_pred, color='red', linewidth=2)

plt.xlabel('CNN Predictions')
plt.ylabel('y_train')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```
