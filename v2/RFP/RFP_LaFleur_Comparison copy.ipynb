```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import dataset into a pandas data frame

predictions_df = pd.read_csv('../predictions.csv')
predictions_df.head()
```

```python
df = pd.read_csv('2024_06_27_relative_fluorescence_pLac_low_high.csv')
df['log(Relative Fluorescence)'] = np.log(df['Relative Fluorescence'])
df['log(log(Relative Fluorescence))'] = np.log(np.log(df['Relative Fluorescence']))
df.head(None)
```

```python
df.loc[0:8, 'Observed log(TX/Txref)'] = predictions_df.loc[3827, 'Observed log(TX/Txref)'] # not exactly pLac, but matches -10, -35
df.loc[9:17, 'Observed log(TX/Txref)'] = predictions_df.loc[10945, 'Observed log(TX/Txref)']
df.loc[18:26, 'Observed log(TX/Txref)'] = predictions_df.loc[12954, 'Observed log(TX/Txref)']
df.head(None)
```

```python
plt.figure(figsize=(10, 5))

# Use XColumn as x-axis and YColumn as y-axis
plt.scatter(df['Observed log(TX/Txref)'], df['Relative Fluorescence'])

# Add titles and labels
plt.xlabel('Observed log(TX/Txref)')
plt.ylabel('Relative Fluorescence')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the color mapping
color_mapping = {'LB': 'red', 'PBS': 'blue', 'M9': 'green'}

# Map the "Media" column to colors
colors = df['Media'].map(color_mapping)

# Create the scatter plot
plt.figure(figsize=(10, 5))
for media, color in color_mapping.items():
    subset = df[df['Media'] == media]
    plt.scatter(subset['Observed log(TX/Txref)'], subset['log(Relative Fluorescence)'], c=color, label=media)

plt.xlabel('Observed log(TX/Txref)')
plt.ylabel('log(Relative Fluorescence)')
plt.legend(title='Media')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Observed log(TX/Txref)'], df['log(log(Relative Fluorescence))'])
plt.xlabel('Observed log(TX/Txref)')
plt.ylabel('log(Relative Fluorescence)')
plt.show()
```
