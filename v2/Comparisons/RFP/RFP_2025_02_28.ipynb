```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../Data/RFP/2025_02_28_v1.csv')
df.head()
```

```python
# plot 'Strain' vs 'Rel_RFP'
plt.figure(figsize=(10, 6))
plt.plot(df['Strain'], df['Rel_RFP'], marker='o', linestyle='-')
plt.title('RFP Data: Strain vs Rel_RFP')
plt.xlabel('Strain')
plt.ylabel('Rel_RFP')
plt.tight_layout()
plt.show()

# plot 'Strain' vs 'Rel_GFP'
plt.figure(figsize=(10, 6))
plt.plot(df['Strain'], df['Rel_GFP'], marker='o', linestyle='-')
plt.title('RFP Data: Strain vs Rel_GFP')
plt.xlabel('Strain')
plt.ylabel('Rel_GFP')
plt.tight_layout()
plt.show()
```
