```python
import random
import numpy as np
import pandas as pd

from function_modules.evaluation_module import *
```

```python
# Shared directory to save results 
shared_data_dir = 'Data/combinatorial_comparison/by_element/'

# Define shared testing parameters for the combinatorial and genetic algorithm
kwargs = {
    'cnn_model_path': '../../Models/CNN_6_1_2.keras',
    'seed': 0,
}
target_expressions = [0, 0.5, 1]
lineages = 10

# Define the shared testing parameters for the combinatorial algorithm
seed = 0
random.seed(seed)
np.random.seed(seed)

# Generate the masked sequences based on pLac elements
pLac = {
    'UP_1' : 'AATACTAGAG',
    'UP_2' : 'GTCTTCCGAC',
    'h35' : 'TTTACA',
    'spacs_1' : 'CTTTATGCT',
    'spacs_2' : 'TCCGGCTCG',
    'h10' : 'TATGTT',
    'disc' : 'GTGTGGGC',
    'ITR_1' : 'GGGAAGAC',
    'ITR_2' : 'AACTAGGGG',
}
order = ['UP_1', 'UP_2', 'h35', 'spacs_1', 'spacs_2', 'h10', 'disc', 'ITR_1', 'ITR_2']
to_mask = ['UP_1', 'UP_2', 'h35', 'spacs_1', 'spacs_2', 'h10', 'disc', 'ITR_1', 'ITR_2']

masked_sequences = {}
for element in to_mask:
    i = order.index(element)
    up = ''.join([pLac[order[j]] for j in range(i)])
    mask = 'N' * len(pLac[order[i]])
    down = ''.join([pLac[order[j]] for j in range(i+1, len(order))])
    masked_sequences[element] = up + mask + down
```

```python
CA_df = test_combinatorial(masked_sequences, target_expressions, kwargs, to_csv=f'{shared_data_dir}combinatorial_results.csv')
```

```python
GA_df = test_genetic(masked_sequences, target_expressions, lineages, kwargs, to_csv=f'{shared_data_dir}genetic_results.csv')
```
