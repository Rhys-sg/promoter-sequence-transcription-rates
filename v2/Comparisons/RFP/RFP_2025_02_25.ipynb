```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
RFP_df = pd.read_csv('../../Data/RFP/2025_02_25.csv')
RFP_df.head()
```

```python
# plot Strain vs RelRFP, with different colors for each media
plt.figure(figsize=(10, 6))
for media, group in RFP_df.groupby('Media'):
    plt.scatter(group['Strain'], group['RelRFP'], label=media)
plt.xlabel('Relative RFP')
plt.ylabel('Strain')
plt.yscale('log')
plt.title('Strain vs Relative RFP')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# plot Strain vs RelRFP for each media separately
for media, group in RFP_df.groupby('Media'):
    plt.scatter(group['Strain'], group['RelRFP'])
    plt.xlabel('Relative RFP')
    plt.ylabel('Strain')
    plt.yscale('log')
    plt.title(f'Strain vs Relative RFP in {media}')
    plt.tight_layout()
    plt.show()
```

```python
# plot Strain vs RelRFP, filter out p625, p750, p875

filtered_df = RFP_df[~RFP_df['Strain'].isin(['p625', 'p750', 'p875'])]

plt.figure(figsize=(10, 6))
for media, group in filtered_df.groupby('Media'):
    plt.scatter(group['Strain'], group['RelRFP'], label=media)
plt.xlabel('Relative RFP')
plt.ylabel('Strain')
plt.yscale('log')
plt.title('Strain vs Relative RFP')
plt.legend()
plt.tight_layout()
plt.show()
```
