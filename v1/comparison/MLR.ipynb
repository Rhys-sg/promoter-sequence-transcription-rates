## Multiple Linear Regression
This approach treats every unique sequaence in each column as a seperate catagorical input. With 5 rows and 11942 columns, fitting a MLR model to this data would take an unrealistic amount of time and be very inaccurate. for the code to run this, see "MLR.ipynb".

```python
import pandas as pd
import numpy as np
```

```python
# import dataset into a pandas data frame

df = pd.read_csv('41467_2022_32829_MOESM5_ESM.csv')
df.head()
```

```python
# All input and output data

X = df[['UP', 'h35', 'spacs', 'h10', 'disc', 'ITR']]
y = df['Observed log(TX/Txref)']

X.head()
```

```python
# one-hot encode the each unique entry in each column
import statsmodels.api as sm

X = sm.add_constant(X[X.columns])
X = pd.get_dummies(X, columns=X.columns, drop_first=True)
X.head()
```

```python
# Label encoding to change "True"/"False" to binary
from sklearn.preprocessing import LabelEncoder

label_columns = X.columns
label_encoder = LabelEncoder()

for col in label_columns:
    X[col] = label_encoder.fit_transform(X[col])

X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.2, random_state = 0)

# Fit the multiple linear regression model
model = sm.OLS(y_train, X_train).fit()

y_pred = model.predict(X_test)

print(f"MSE: {mean_squared_error(X_test, y_pred)}")
```
