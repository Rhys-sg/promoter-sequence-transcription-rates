```python
import pandas as pd
import random
```

```python
df = pd.read_csv('../../Data/LaFleur/41467_2022_32829_MOESM5_ESM.csv')[['UP', 'h35', 'spacs', 'h10', 'disc', 'ITR']]
```

```python
# Get sorted lists of entries for each column
sorted_entries = {}
for column in df.columns:
    sorted_entries[column] = df[column].value_counts().index.tolist()

# Output the sorted entries for each column
for column, entries in sorted_entries.items():
    print(f"{column}: {entries}")
```

```python
known_sequences = {(data['UP'], data['h35'], data['spacs'], data['h10'], data['disc'], data['ITR']) for _, data in df.iterrows()}
```

```python
# Generate X sequences not in known_sequences
def generate_sequences(df, sorted_entries, known_sequences, X, seed=None):
    generated_sequences = set()
    
    # Set the random seed if provided
    if seed is not None:
        random.seed(seed)

    while len(generated_sequences) < X:
        new_sequence = tuple(
            random.choice(sorted_entries[column][::-1]) for column in df.columns
        )
        
        if new_sequence not in known_sequences:
            generated_sequences.add(new_sequence)
        
    return pd.DataFrame(list(generated_sequences), columns=df.columns)

# Generate 5 new sequences not in known_sequences
generated_sequences_df = generate_sequences(df, sorted_entries, known_sequences, 100, seed=42)
generated_sequences_df.head()
# generated_sequences_df.to_csv('generated_sequences.csv', index=False)
```
