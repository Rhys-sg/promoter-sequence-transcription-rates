```python
# Shared directory to save results 
shared_data_dir = 'Data/parameter_testing/selection/'

# Shared GA parameters
kwargs = {
    'cnn_model_path': '../../Models/CNN_6_1_2.keras',
    'masked_sequence': 'AATACTAGAGGTCTTCCGACNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNGTGTGGGCGGGAAGACAACTAGGGG',
    'seed': 0,
}

# Testing parameters
target_expressions = [0, 0.5, 1]
lineages = 10
test_cases = 11
```

```python
# Read results (if necessary) and plot
results_df = pd.read_csv(f'{shared_data_dir}selTournament.csv')

for target_expression in target_expressions:
    filtered_df = results_df[results_df['target_expression'] == target_expression]
    scatter_plot(filtered_df, target_expression, 'tournsize', polynomial_degree=2)
```

```python
# Evaluate the selection methods without additional parameters
param_ranges = {
    'selection_method' : ['selRandom', 'selBest', 'selRoulette', 'selStochasticUniversalSampling', 'selNormRoulette'],
}
no_param_df = test_params(param_ranges, target_expressions, lineages, kwargs)

# Evaluate selTournament
param_ranges = {
    'selection_method' : ['selTournament'],
    'tournsize' : [2],
}
selTournament_df = test_params(param_ranges, target_expressions, lineages, kwargs)

# Evaluate selBoltzmann
param_ranges = {
    'selection_method' : ['selBoltzmann'],
    'boltzmann_temperature' : [0],
}
selBoltzmann_df = test_params(param_ranges, target_expressions, lineages, kwargs)

# Evaluate selTournamentWithoutReplacement
param_ranges = {
    'selection_method' : ['selTournamentWithoutReplacement'],
    'tournsize' : [2],
}
selTournamentWithoutReplacement_df = test_params(param_ranges, target_expressions, lineages, kwargs)

# Concatenate all results, and save to csv
results_df = pd.concat([no_param_df, selTournament_df, selBoltzmann_df, selTournamentWithoutReplacement_df])
results_df.to_csv(f'{shared_data_dir}selection_comparison.csv', index=False)
```
