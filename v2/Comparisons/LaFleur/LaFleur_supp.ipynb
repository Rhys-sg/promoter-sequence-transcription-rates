```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = '../../Data/LaFleur/41467_2022_32829_MOESM5_ESM.xlsx'

all_sheet_names = pd.ExcelFile(file_path).sheet_names

slice = all_sheet_names[3:10] + all_sheet_names[13:15]

sheets = pd.read_excel(file_path, sheet_name=slice)

sheet_names = list(sheets.keys())
```

```python
sheet_names
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

promoter_lengths = []

for name, df in sheets.items():
    for i, row in df.iterrows():
        sequence = ''
        if 'Upstream DNA' in df.columns:
            sequence += row['Upstream DNA']
        sequence += row['Promoter Sequence']
        if 'Downstream DNA' in df.columns:
            sequence += row['Downstream DNA']
        promoter_lengths.append({'Sheet': name, 'Length': len(sequence)})
```

```python
df_lengths = pd.DataFrame(promoter_lengths)
sns.boxplot(x='Length', y='Sheet', data=df_lengths)
plt.title('Distribution of Promoter Lengths by Sheet')
plt.show()
```

```python
combined_lengths = df_lengths['Length']
sns.boxplot(x=combined_lengths)
plt.title('Distribution of Promoter Lengths by Sheet')
plt.xticks(rotation=90)
plt.show()
```
