```python
import numpy as np
from GA_params_test import GeneticAlgorithm
```

```python
cnn_model_path = '../../Models/CNN_6_1_2.keras'
masked_sequence = 'AATACTAGAGGTCTTCCGACNNNNNNNNNNNNNNNNNNNNNNNNNNNNNNGTGTGGGCGGGAAGACAACTAGGGG'
```

```python
def get_sequences(target_expression):
    sequences = []
    predictions = []
    for i in range(5):
        ga = GeneticAlgorithm(
            cnn_model_path=cnn_model_path,
            masked_sequence=masked_sequence,
            target_expression=target_expression,
            pop_size=300,
            generations=100,
            seed=i,
            print_progress=False
        )
        best_sequences, best_predictions = ga.run()
        sequences.append(best_sequences)
        predictions.append(best_predictions)
    
    sorted_results = sorted(
        zip(sequences, predictions),
        key=lambda x: abs(x[1] - target_expression)
    )
    
    for sequence, prediction in sorted_results:
        print(f'{sequence} : {prediction:.3f}')
```

```python
for target_expression in np.arange(0, 1.125, 0.125):
    print(f'Target expression: {target_expression}')
    get_sequences(target_expression)
    print()
```
