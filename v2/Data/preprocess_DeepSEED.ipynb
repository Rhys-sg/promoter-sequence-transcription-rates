# Description

Truncate the promoters from  Zhang (2023) DeepSEED to be 150bp and min-max scale the expression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import keras
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
```

#### For ecoli_mpra_3_laco.csv
----

```python
df = pd.read_csv('../Data/Zhang (DeepSEED)/ecoli_mpra_3_laco.csv')
df.head()
```

```python
# truncate the sequences (realA and realB) to be 150bp long
df['realA_trunc'] = df['realA'].str[:150]
df['realB_trunc'] = df['realB'].str[:150]

# min-max scale the values in the dataframe
df['expr_min_max'] = MinMaxScaler().fit_transform(df['expr'].values.reshape(-1, 1))

df.head()
```

```python
# save the dataframe to a new csv file
df.to_csv('../Data/Zhang (DeepSEED)/ecoli_mpra_3_laco_processed.csv', index=False)
```
